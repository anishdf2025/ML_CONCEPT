# üìò Recurrent Neural Network (RNN)

1. Definition

A Recurrent Neural Network (RNN) is a type of neural network specifically designed to handle sequential or time-dependent data.
Unlike feedforward networks, RNNs contain recurrent (feedback) connections that allow information from previous time steps to influence the current output.
This gives RNNs the ability to remember past information through an internal state called the hidden state.

Keywords:
Sequential data, temporal dependency, hidden state, memory, recurrence.

2. Why RNN? (Motivation)

Traditional neural networks assume:

All inputs are independent.

But in real-world problems, order matters.

Examples:

In a sentence, meaning depends on earlier words.
(‚Äúbank deposit cash‚Äù vs ‚Äúriver bank‚Äù)

Stock price at time t depends on previous prices.

Audio signals depend on earlier sound waves.

RNNs solve this by keeping a memory of past inputs, enabling them to model temporal patterns.

## **3. RNN Architecture**

RNN is a neural network designed to handle **sequential data** by maintaining a **hidden state (memory)** across time steps.

### **RNN consists of:**
- **Input sequence:**  
  \( x_1, x_2, x_3, ‚Ä¶, x_T \)

- **Hidden state (memory):**  
  \( h_t \)

- **Output sequence:**  
  \( y_t \)

---

## **How RNN Works at Time Step t**

At each step \( t \):

1. Takes current input \( x_t \)  
2. Takes previous hidden state \( h_{t-1} \)  
3. Combines them to produce new hidden state \( h_t \)  
4. Produces output \( y_t \)

---

## **3.1 Mathematical Formulation**

### **Hidden State Update**
\[
h_t = \tanh(W_x x_t + W_h h_{t-1} + b)
\]

### **Output Calculation**
\[
y_t = W_y h_t
\]

---

## **Where:**

| Symbol | Meaning |
|-------|---------|
| \( W_x \) | Input weight matrix |
| \( W_h \) | Recurrent weight matrix |
| \( W_y \) | Output weight matrix |
| \( b \) | Bias term |
| \( \tanh \) | Activation function (keeps values between ‚àí1 to 1) |
