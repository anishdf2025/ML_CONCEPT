```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Linear Activation Function
# -----------------------------
def linear(x):
    return x

# Generate values
x = np.linspace(-10, 10, 400)
y = linear(x)

# Plotting
plt.figure(figsize=(7,5))
plt.plot(x, y, label='f(x) = x', linewidth=1)

plt.title("Linear Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Output f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x):
    return 1/(1+np.exp(-x))
x=np.linspace(-10,10,500)
y=sigmoid(x)
plt.figure(figsize=(7,5))
plt.plot(x,y,label='(1/(1+np.exp(-x)))',linewidth=2)
plt.title("Sigmoid activation function")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def Relu_activation(x):
    return np.maximum(0,x)
x=np.linspace(-10,10,500)
y=Relu_activation(x)
plt.figure(figsize=(7,5))
plt.plot(x,y,label='ReLU',linewidth=2)
plt.title("ReLU Activation function")
plt.xlabel("x-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def Leaky_Relu_activation(x):
    return np.maximum(0.1*x,x)
x=np.linspace(-10,10,500)
y=Leaky_Relu_activation(x)
plt.figure(figsize=(7,5))
plt.title("LeakyReLU Activation function")
plt.plot(x,y,label='LeakyRelu',linewidth=2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt
def tan(x):
    return np.tanh(x)
x=np.linspace(-10,10,500)
y=tan(x)
plt.figure(figsize=(7,5))
plt.plot(x,y,label='Tanh',linewidth=2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```
